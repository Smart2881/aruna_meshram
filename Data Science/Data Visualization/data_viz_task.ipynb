```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load data
df = pd.read_csv('Cars93.csv')
# Show columns
df.columns
```

### 1. Boxplot

```python
# Boxplot for revs per mile

cars = df.loc[df['Manufacturer'].isin(['Audi','Hyundai','Suzuki','Toyota'])]
sns.boxplot(data=cars,x='Manufacturer', y='Rev.per.mile',hue='Manufacturer' ,palette='pastel')
plt.title('Rev per mile by Manufacturer')
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?

**Boxplot answer:**  
Suzuki has highest revs per mile.

### 2. Histogram

```python
# Histogram of MPG
sns.set_theme(style="ticks")
plt.title('MPG in City & Highway')

sns.histplot(data=cars, x='MPG.city', color='orange', label='MPG.city', kde=True)
sns.histplot(data=cars, x='MPG.highway', color='pink', label='MPG.highway', kde=True)

plt.legend()
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**  
Looking at the barplot, MPG on highway is more than city road.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
sns.lineplot(x='Wheelbase', y='Turn.circle', data=cars, color='blue')
plt.title('Wheelbase vs Turning Circle')
plt.xlabel('Wheelbase')
plt.ylabel('Turning Circle')
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?

**Lineplot answer:**  
The relationship between Turning circle and Wheelbase is, higher the wheelbase the larger is the turning circle.

### 4. Barplot

```python
# Barplot for mean horsepower
sns.barplot(data=cars, x='Type', y='Horsepower', hue="Type", legend=False)
plt.title('Mean Horsepower by Car Type')
plt.xlabel('Car Type')
plt.ylabel('Mean Horsepower')
```

**Barplot question:**  
Does a larger car mean more horsepower?

**Barplot answer:**  
No, large size car does not mean more horsepower. The above barplot shows compact car has highest horsepower compared to Van
